# Bell State Deep Dive

A closer look at the Bell state circuit: inspection, code generation, and simulation.

```python
from qdk_pythonic import Circuit
```

## Build the Circuit

```python
circ = Circuit()
q = circ.allocate(2, label="bell")

circ.h(q[0]).cx(q[0], q[1]).measure_all()
```

## Inspect the Circuit

Use `draw()` to get an ASCII diagram, and check `gate_count()` and `depth()`.

```python
print(circ.draw())
```

```python
print(f"Gate count: {circ.gate_count()}")
print(f"Depth:      {circ.depth()}")
print(f"Qubits:     {circ.qubit_count()}")
```

## Q# and OpenQASM Output

The circuit can be exported to both Q# and OpenQASM 3.0 source strings.

```python
print("=== Q# ===")
print(circ.to_qsharp())
print()
print("=== OpenQASM 3.0 ===")
print(circ.to_openqasm())
```

## Simulate the Circuit

> **Note:** Running simulation requires the `qsharp` package (`pip install qsharp>=1.25`).

```python
# Requires: pip install qsharp>=1.25
results = circ.run(shots=100)
print(results)
```
